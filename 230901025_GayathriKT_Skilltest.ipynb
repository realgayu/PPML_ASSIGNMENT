```python
#1A
import statistics

exam_scores = [78, 65, 92, 88, 55, 72, 98, 60, 80, 75, 58, 82, 95, 70, 68, 85, 77, 63, 90, 89]

average_score = statistics.mean(exam_scores)
highest_score = max(exam_scores)
lowest_score = min(exam_scores)

print(f"Average Score: {average_score:.2f}")
print(f"Highest Score: {highest_score}")
print(f"Lowest Score: {lowest_score}")

pass_mark = 50
passed_students = [score for score in exam_scores if score >= pass_mark]
failed_students = [score for score in exam_scores if score < pass_mark]

print(f"\nNumber of students who passed: {len(passed_students)}")
print(f"Number of students who failed: {len(failed_students)}")

def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    elif score >= 50:
        return "E"
    else:
        return "F"

print("\nStudent Grades:")
for score in exam_scores:
    grade = get_grade(score)
    print(f"Score: {score}, Grade: {grade}")
```

```python
#3A

import cmath
import pandas as pd
def imp(c,v,f):
  w=2*cmath.pi*f
  if c=='R':
    return complex(v,0)
  elif c=='L':
    return complex(0,w*v)
  elif c=='C':
    return complex(0,-1/(w*v))
  else:
    return 'Invalid component'
def tot(df,config):
  if config=='Series':
    return df['Impedance'].sum()
  else:
    return 1 / sum(1 / z for z in df['Impedance'])
f=float(input("Enter Frequency:"))
config=input("Enter the configuration(Series/Parallel):")
n=int(input('No of components:'))
d=[]
for i in range(n):
  c=input("Enter the component(R/L/C):").upper()
  v=float(input("Enter the value:"))
  z=imp(c,v,f)
  d.append({'Component':f'{c}{i+1}','Type':c,'Value': v, 'Impedance': z})
df=pd.DataFrame(d)
pd.set_option("display.precision",2)
print(df)
t=tot(df,config)
print(f'Total Impedance:{t:3.2f}')
```

```python
#4A

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sampling_rate = 50  # Hz
duration = 1  # in seconds
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
def generate_wave(freq, amp, noise_level):
    signal = amp * np.sin(2 * np.pi * freq * t)
    noise = np.random.normal(0, noise_level, t.shape)
    return signal + noise
data={
    'Patient_ID':[f"{i}" for i in range(1,51)],
    'Time(s)':t,
    'Delta': generate_wave(2, 10, 10),
    'Theta': generate_wave(6, 6, 8),
    'Alpha': generate_wave(10, 4, 5),
    'Beta':  generate_wave(20, 5, 3)
}
df = pd.DataFrame(data)
df.to_csv(r"C:\Users\Gayathri\Documents\skilltestsamplefile.csv",index=False)
print(df)
print("\nStatistics (Mean, Variance, Max):")
print('Mean:\n',df[['Delta', 'Theta', 'Alpha', 'Beta']].mean())
print('Variance:\n',df[['Delta', 'Theta', 'Alpha', 'Beta']].var())
print('Max:\n',df[['Delta', 'Theta', 'Alpha', 'Beta']].max())
plt.bar(df['Patient_ID'],df['Delta'])
plt.xlabel('Patient_ID')
plt.ylabel('Delta')
plt.show()
plt.bar(df['Patient_ID'],df['Theta'])
plt.xlabel('Patient_ID')
plt.ylabel('Theta')
plt.show()
plt.bar(df['Patient_ID'],df['Alpha'])
plt.xlabel('Patient_ID')
plt.ylabel('Alpha')
plt.show()
plt.bar(df['Patient_ID'],df['Beta'])
plt.xlabel('Patient_ID')
plt.ylabel('Beta')
plt.show()
```

```python
#4B
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}

df = pd.DataFrame(data)


print("1. Inventory Data:")
print(df)

df['Total_Value'] = df['Price'] * df['Quantity']
print("\n2. Data with Total Value:")
print(df)


most_expensive = df.loc[df['Price'].idxmax()]
print("\n3. Most Expensive Product:")
print(most_expensive)

total_inventory_value = df['Total_Value'].sum()
print("\n4. Total Inventory Value:", total_inventory_value)


high_stock = df[df['Quantity'] > 20]
print("\n5. Products with quantity > 20:")
print(high_stock)


plt.figure(figsize=(8, 5))
plt.bar(df['Product'], df['Total_Value'], color='skyblue')
plt.title('Total Value of Inventory per Product')
plt.xlabel('Product')
plt.ylabel('Total Value ($)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
